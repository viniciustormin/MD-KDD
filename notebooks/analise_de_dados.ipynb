### Teste para checar se o diagnóstico é dependente do sexo

```python
df2 = df1.copy()

# Normalizar valores de Sexo e Diagnóstico
df2['SEXO'] = df2['SEXO'].str.strip()
df2['SEXO'] = df2['SEXO'].replace({
    'F': 'Feminino', 
    'M': 'Masculino',
    'masculino': 'Masculino'
})

df2['NORMAL X ANORMAL'] = df2['NORMAL X ANORMAL'].str.strip()
df2['NORMAL X ANORMAL'] = df2['NORMAL X ANORMAL'].replace({
    'anormal': 'Anormal',
    'Normais': 'Normal'
})

# Análise de correlação cruzada: Sexo x Diagnóstico
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico 1: Contagem absoluta
crosstab_counts = pd.crosstab(df2['SEXO'], df2['NORMAL X ANORMAL'])
ax0 = crosstab_counts.plot(
    kind='bar', 
    ax=axes[0], 
    color=['#FF6B6B', '#4ECDC4'],
    edgecolor='black',
    width=0.7
)
axes[0].set_title('Distribuição de Diagnóstico por Sexo (Contagem)')
axes[0].set_xlabel('Sexo')
axes[0].set_ylabel('Quantidade')
axes[0].legend(title='Diagnóstico', loc='upper right')
axes[0].grid(axis='y', alpha=0.3)
axes[0].tick_params(axis='x', rotation=45)

# Gráfico 2: Proporção normalizada (%)
crosstab_prop = pd.crosstab(df2['SEXO'], df2['NORMAL X ANORMAL'], normalize='index')
ax1 = crosstab_prop.plot(
    kind='bar', 
    ax=axes[1], 
    color=['#FF6B6B', '#4ECDC4'],
    edgecolor='black',
    width=0.7
)
axes[1].set_title('Proporção de Diagnóstico por Sexo (%)')
axes[1].set_xlabel('Sexo')
axes[1].set_ylabel('Proporção')
axes[1].legend(title='Diagnóstico', loc='upper right')
axes[1].grid(axis='y', alpha=0.3)
axes[1].tick_params(axis='x', rotation=45)
axes[1].yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y*100:.0f}%'))

plt.tight_layout()
plt.show()

# Tabela de contingência com percentuais
print("TABELA DE CONTINGÊNCIA - SEXO X DIAGNÓSTICO:\n")
tabela = pd.crosstab(df2['SEXO'], df2['NORMAL X ANORMAL'], margins=True)
print(tabela)
print("\n\nPROPORÇÕES POR SEXO (%):\n")
proporcoes = pd.crosstab(df2['SEXO'], df2['NORMAL X ANORMAL'], normalize='index') * 100
print(proporcoes.round(2))
```

Então concluímos que não é preciso normalizar as duas colunas, pois a doença não é especifica para um genêro
